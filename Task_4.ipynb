```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\navee\Downloads\twitter_dataset.csv")
```

```python
data.head()
```

```python
data.Sentiment.value_counts()
```

```python
data = data[data['Sentiment'].isin(['Positive','Negative','Neutral'])]
```

```python
data['Sentiment'].value_counts().plot(kind='bar')
plt.xlabel("Sentiment")
plt.ylabel("Value")
plt.title("Sentiment Distribution")
```

```python
data['Sentiment'].value_counts().plot(kind='pie')
plt.title("Sentiment Distribution")
```

- Therefore, using all these results we can conclude that there were more number of negative feedbacks over the positive ones
- However, the difference between them is negligible indicating that the app had an overall neutral effect on people
